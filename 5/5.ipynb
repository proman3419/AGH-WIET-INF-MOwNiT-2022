```python
import numpy as np
import matplotlib.pyplot as plt
import time

def time_exec(func, *args):
    start_time = time.time()
    result = func(*args)
    return result, time.time() - start_time
```

## Zadanie 1

```python
ax = plt.figure().add_subplot(projection='3d')

s, t = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
x = np.cos(s) * np.sin(t)
y = np.sin(s) * np.sin(t)
z = np.cos(t)
ax.plot_surface(x, y, z)

plt.show()
```

```python
A = np.random.random((3, 3))
AA = np.random.random((3, 3))
AAA = np.random.random((3, 3))
AAAA = np.matmul(np.matmul(A, AA), AAA)

sphere = np.array([np.column_stack((x[i],y[i],z[i])) @ AAAA for i in range(len(x))])
_x = sphere[:,:,0]
_y = sphere[:,:,1]
_z = sphere[:,:,2]

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(_x, _y, _z, label='sphere')
plt.show()
```
